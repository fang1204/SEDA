```python
def is_continuous_from_string(input_str):
    # 將輸入字串轉換為整數列表
    numbers = list(map(int, input_str.split(',')))
    
    # 將數字列表轉換成區間，假設列表中的數字是按區間順序排列
    intervals = [(numbers[i], numbers[i+1]) for i in range(0, len(numbers), 2)]

    # 按照每個區間的起始點排序
    intervals.sort(key=lambda x: x[0])
    # print(intervals)
    # 檢查是否連續
    for i in range(len(intervals) - 1):
        if intervals[i][1]+2 <= intervals[i + 1][0]:
            return True
    return False


import pandas as pd
data = pd.read_csv("D:/hw/W2NER-final/acl/share13/test.ann", names=["file_id", "label", "start_end", "entity", "other"], keep_default_na=False, na_values=['NaN', 'null'], delimiter='\t')
# data = pd.read_csv("/data/m11115088/W2NER_final/acl/share14/test.ann", names=["file_id", "label", "start_end", "entity", "other"], keep_default_na=False, na_values=['NaN', 'null'], delimiter='\t')
# data = pd.read_csv("D:/hw/W2NER-final/acl/cadec/ann", names=["file_id", "label", "start_end", "entity", "other"], keep_default_na=False, na_values=['NaN', 'null'], delimiter='\t')
data["discontinue"] = data['start_end'].apply(lambda x: is_continuous_from_string(x))
data = data[data['discontinue']]
# with open("D:/hw/W2NER-final/answer/cadec/id/test.id") as f:
with open("D:/hw/W2NER-final/answer/share13/id/test.id") as f:
# # with open("/data/m11115088/W2NER_final/answer/share14/id/test.id") as f:
    filename = f.read().split("\n")

test_discontinue = data[data['file_id'].apply(lambda x:x in filename)]
print(len(test_discontinue),len(filename))
test_discontinue
```
